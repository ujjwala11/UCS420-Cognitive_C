Q.1 For the array gfg = np.matrix('[4, 1, 9; 12, 3, 1; 4, 5, 6]'), Find i. Sum of all elements ii. Sum of all elements row-wise iii. Sum of all elements column-wise

```python
import numpy as np
gfg = np.matrix('[4, 1, 9; 12, 3, 1; 4, 5, 6]')

# i. Sum of all elements
total_sum = np.sum(gfg)

# ii. Sum of all elements row-wise
row_sum = np.sum(gfg, axis=1)

# iii. Sum of all elements column-wise
column_sum = np.sum(gfg, axis=0)

print("Sum of all elements:", total_sum)
print("Sum of elements row-wise:", row_sum)
print("Sum of elements column-wise:", column_sum)
```

Q.2 (a)For the array: array = np.array([10, 52, 62, 16, 16, 54, 453]), find i. Sorted array ii. Indices of sorted array iii. 4 smallest elements iv. 5 largest elements

```python
import numpy as np

array = np.array([10, 52, 62, 16, 16, 54, 453])

sorted_array = np.sort(array)
sorted_indices = np.argsort(array)
smallest_elements = np.sort(array)[:4]
largest_elements = np.sort(array)[-5:]

print("Sorted array:", sorted_array)
print("Indices of sorted array:", sorted_indices)
print("4 smallest elements:", smallest_elements)
print("5 largest elements:", largest_elements)
```

(b) For the array: array = np.array([1.0, 1.2, 2.2, 2.0, 3.0, 2.0]), find i. Integer elements only ii. Float elements only

```python
import numpy as np

array = np.array([1.0, 1.2, 2.2, 2.0, 3.0, 2.0])

integer_elements = array[np.floor(array) == array]
float_elements = array[np.floor(array) != array]

print("Integer elements only:", integer_elements)
print("Float elements only:", float_elements)
```

Q.3 You are given a weekly sales dataset and need to perform various operations using NumPy broadcasting. a) Generate your unique sales dataset:  Take the sum of the ASCII values of the initials of your first and last name. Call this value X. (If your initials are A B → ASCII sum = 65


* 66 = 131 → sales = [131, 181, 231, 281, 331].)  Create a NumPy array sales with values [X, X+50, X+100, X+150, X+200]. b) Compute your personalized tax rate as ((X % 5) + 5) / 100.  Use broadcasting to apply this tax rate to each sales value. c) Adjust sales based on discount:  If sales < X+100, apply a 5% discount.  If sales >= X+100, apply a 10% discount. d) Expand sales data for multiple weeks:  Create a 3×5 matrix representing three weeks of sales by stacking sales three times using broadcasting.  Increase sales by 2% per week using element-wise broadcasting.




```python
import numpy as np

# Step 1: Compute X and Sales Array
X = ord('R') + ord('G')  # ASCII sum
sales = np.array([X, X+50, X+100, X+150, X+200])

# Step 2: Compute Tax Rate and Apply
tax_rate = ((X % 5) + 5) / 100
tax_amount = sales * tax_rate

# Step 3: Apply Discount
discounted_sales = np.where(sales < X+100, sales * 0.95, sales * 0.90)

# Step 4: Expand Sales Data for Multiple Weeks
weeks = 3
weekly_sales = np.vstack([discounted_sales * (1 + i * 0.02) for i in range(weeks)])

# Print results
print("Sales Array:", sales)
print("Tax Rate:", tax_rate)
print("Tax Amount:", tax_amount)
print("Discounted Sales:", discounted_sales)
print("Weekly Sales:\n", weekly_sales)
```
